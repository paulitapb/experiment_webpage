Generate data.js based on curent imgs

```python
import json
import os
import pandas as pd
```

```python
experiment_files = os.listdir("client/public/images/generated_imgs/")    
experiment_files_dir = {
    'SD-english-LORA1': 'experiment_english_captions_LORA',
    'SD-english-txtInv': 'experiment_english_captions_textual_inversion',
    'SD-spanish': 'experiment_spanish_captions',
    'SD-english': 'experiment_english_captions',
    'SD-english-LORA2' : 'experiment_english_caps_LORA_V1_intento2',
}
```

```python
data = []
id = 0
for exp in experiment_files:
    images = os.listdir(f"client/public/images/generated_imgs/{exp}")
    
    for image in images:
        image_info = image.split('_')
        
        group = int(image_info[1].replace('group', ''))
        
        img = int(image_info[2].replace('img', ''))
        promtUsed = int(image_info[0].replace('index', ''))
        
        data.append({
            'id': id,
            'name': f'img generada de grupo {group} img {img}',
            'dir': f'/images/generated_imgs/{exp}/{image}',
            'alt': '',
            'img': img,
            'group': group,
            'imgGeneratedBy': exp, 
            'promptUsed': promtUsed
            })
        id += 1

with open('client/src/data.json', 'w') as f:
    json.dump(data, f, indent=2)
```

```python
experiment_files = os.listdir("client/public/images/generated_imgs/")        
experiment_files = [f for f in experiment_files if (not f.startswith('dibujos') and not f.startswith('DALLE'))]
experiment_files
```

```python
image_data = pd.read_json('../selected_images_piloto_fam4.json')
image_data
```

```python
data = []

for exp in experiment_files:
    images = os.listdir(f"client/public/images/generated_imgs/{exp}")
    
    for image in images:
        img_name, _ = os.path.splitext(image)
        image_info = img_name.split('_')
        
        group = int(image_info[1].replace('group', ''))
        
        img = int(image_info[2].replace('img', ''))
        promtUsed = int(image_info[0].replace('index', ''))
        
        
        img_dir = image_data[image_data['dir'].str.contains(experiment_files_dir[exp]) & image_data['dir'].str.contains(img_name)]


        id = img_dir['id'].values[0]
        data.append({
            'id': int(id),
            'name': f'img generada de grupo {group} img {img}',
            'dir': f'/images/generated_imgs/{exp}/{image}',
            'alt': '',
            'img': img,
            'group': group,
            'imgGeneratedBy': exp, 
            'promptUsed': promtUsed
            })

dib_prompt_data = pd.read_csv('../repo-fine-tuning/data_past_experiment/train_captions_joinIdPrompts.csv')

for dib in os.listdir("client/public/images/generated_imgs/dibujos"):
    
    dib, _ = os.path.splitext(dib)
    dib_cap = dib.split('cap')[1]

    prompt_info = dib_prompt_data[dib_prompt_data['originalIdCaption'] == int(dib_cap)]
    promptUsed = prompt_info['idPrompt'].values[0]
 
    group = int(prompt_info['group'].values[0])
    
    img = int(prompt_info['img_num'].values[0])

    
    img_dir = image_data[image_data['dir'].str.contains('dibujos') & image_data['dir'].str.contains(dib)]

    id = img_dir['id'].values[0]

    data.append({
        'id': int(id),
        'name': f'img generada de grupo {group} img {img}',
        'dir': f'/images/generated_imgs/dibujos/{dib}.png',
        'alt': '',
        'img': img,
        'group': group,
        'imgGeneratedBy': 'dibujos', 
        'promptUsed': promtUsed
        })
    
for image in os.listdir("client/public/images/generated_imgs/DALLE-english"):
    image_name, _ = os.path.splitext(image)
    image_info = image_name.split('_')

    
    group = int(image_info[1].replace('group', ''))
    
    img = int(image_info[2].replace('img', ''))
    promtUsed = int(image_info[0].replace('index', ''))

    data.append({
        'id': int(promptUsed),
        'name': f'DALLE image',
        'dir': f'/images/generated_imgs/DALLE-english/{image}',
        'alt': '',
        'img': img,
        'group': group,
        'imgGeneratedBy': 'DALLE-english', 
        'promptUsed': promtUsed
        })

with open('client/src/data.json', 'w') as f:
    json.dump(data, f, indent=2)
```
